1. extract the hp tensor from the output of a gaussian file

2. convert the componts to nlo coefficients

3. boltzmann weighting of conformers

Remember that python uses zero indexing 

```python
import hp2nlo
```

```python
tensor = hp2nlo.extract_hyperpolarizability(filename = "rm734_800_1064.log",wavelength=800.0, beta_type="Beta(-2w;w,w)", orientation="input")
```

```python
tensor
```

```python
# convert to coefficients
# experimental constants can be found in hp2nlo
d33, d31, d15 = hp2nlo.convert_hp_2_nlo(tensor=tensor, density=1280, mol_mass = 424, p1_val = 0.723, p3_val= -0.14 )
```
